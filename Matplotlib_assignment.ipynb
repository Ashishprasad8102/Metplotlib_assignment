```python
1>Matplotlib is a powerful library that provides a wide range of functions for creating various types
of plots and charts. It is extensively used in data analysis, scientific research, and visualization
tasks due to its flexibility and ease of use.


Here are five common types of plots that can be created using the pyplot module of Matplotlib:

Line Plot: A simple plot where data points are connected by straight lines, often used to show trends
or time series data.

Scatter Plot: A plot with individual data points represented as dots, useful for visualizing the 
distribution or relationship between two variables.

Bar Plot: A plot where rectangular bars represent the data, commonly used to compare discrete categories.

Histogram: A graphical representation of the distribution of continuous data, divided into bins to 
display frequency or probability density.

Pie Chart: A circular chart that shows the proportions of different categories as wedges, often used
to represent parts of a whole.


Matplotlib offers many more types of plots, including 3D plots, heatmap, box plot, and more. Its
versatility and ability to customize the plots make it a widely-used library for data visualization in Python
```

```python
2>A scatter plot is a type of data visualization used to display the relationship or correlation between
two numerical variables. It is called a "scatter" plot because it scatters individual data points on the
graph.
```

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x=3+np.random.normal(0,2,50)
y=3+np.random.normal(0,2,len(x))
plt.scatter(x,y)
plt.title("scatter chart")
plt.ylabel("y-axixs")
plt.xlabel("x-axixs")
```

```python
3>The subplot() function in Matplotlib is used to create multiple plots (subplots) within a single
figure. It helps in organizing multiple plots on a grid, allowing you to compare different datasets
or visualize various aspects of data side by side.
```

```python
x = [0,1,2,3,4,5,]
x1= [0,1,2,3,4,5]
x2=[0,1,2,3,4,5]
x3 =[0,1,2,3,4,5]

y= [0,100,200,300,400,500]
y1= [50,20,40,20,60,70]
y2=[10,20,30,40,50,60]
y3=[200,350,250,550,450,150]

plt.plot(x1,y1)
plt.plot(x,y)
plt.plot(x2,y2)
plt.plot(x3,y3)
```

```python
4> A bar plot is a type of data visualization that represents data using rectangular bars. It is commonly
used to display and compare the values of different categories or groups. Each bar's length corresponds
to the value of a specific category or group, making it easy to identify relative differences between them.
```

```python
company = ['apple','microsoft','google','AMD']
profit = [3000,8000,1000,10000]
plt.bar(company,profit)
```

```python
5>A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution
of data along a single numerical variable. It provides a summary of the data's central tendency, spread,
and presence of outliers. The plot is based on five summary statistics, which are depicted using visual
elements such as boxes and whiskers.
```

```python
box1 = [100,10,200]
box2 = [90,20,200]
plt.box(box1,box2)
```
